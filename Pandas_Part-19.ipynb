## Pandas Pivote Table

```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'Department': ['HR', 'HR', 'IT', 'IT', 'Sales', 'Sales'],
    'Month': ['Jan', 'Feb', 'Jan', 'Feb', 'Jan', 'Feb'],
    'Revenue': [100, 150, 200, 250, 300, 350]
})
```

```python
df.pivot_table(index='Department',
    columns='Month',
    values='Revenue',
    aggfunc='sum')
```
