## Handling Missing Values in the Python

```python
import pandas as pd
```

```python
df = pd.read_csv(r'F:\Projects\Python Projects\csv\Fortune_10.csv')
df
```

```python
df.dropna()
```

```python
df.dropna(axis=1)
```

```python
df.dropna(how='any')
```

```python
df.dropna(how='all')
```

```python
df.dropna(axis=1,how='any')
```

```python
df.dropna(how='all')
```

```python
df.dropna(thresh=1)
```

```python
df.dropna(subset=['ID'])
```

```python
df.dropna(subset=['Name'])
```

```python
df.dropna(subset=['Profit'])
```

```python
df.dropna(inplace=True)
df
```
