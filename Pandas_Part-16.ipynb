## Pandas Concat() Function #2

```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'ID':[1,2,3,4],'Name':['A','B','C','D'],'Class':[5,6,7,8]})
df1
```

```python
df2 = pd.DataFrame({'ID':[5,6,7,8],'Name':['E','F','G','H'],'Class':[9,10,11,12]})
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],axis=1)
```

```python
pd.concat([df1,df2],axis=1,join='inner')
```

```python
pd.concat([df1, df2.reindex(df1.index)], axis=1)
```

```python
pd.concat([df1, df2.reindex(df2.index)], axis=1)
```

```python
df2 = pd.DataFrame({'ID':[5,6,7,8],'Name':['E','F','G','H'],'Class':[9,10,11,12]})
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],keys = ['df1','df2'])
```

```python
pd.concat([df1,df2],keys = ['First DF','Second DF'])
```

```python
pd.concat([df1,df2],axis=1,keys = ['First DF','Second DF'])
```

```python
df1 = pd.DataFrame({'ID':[1,2,3,4],'Name':['A','B','C','D'],'Class':[5,6,7,8]})
df1
```
